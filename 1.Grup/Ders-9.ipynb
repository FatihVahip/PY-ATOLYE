```python
#kullanıcı adı ve şifresi ile sisteme giriş yapmak
s_k_adi = "Kedi"
s_sifre = "12345"

k_adi = input("Kullanıcı adını gir:")
sifre = input("Şifreni gir:")

if k_adi == s_k_adi and sifre== s_sifre:
    print("Hoşheldin", k_adi)
elif k_adi == s_k_adi and sifre!= s_sifre:
    print("Şifren yanlış, tekrar dene")
elif k_adi != s_k_adi and sifre== s_sifre:
    print("Kullanıcı adın yanlış, tekrar dene")
else:
    print("iki alanı da yanlış girdin:(, tekrar dene ")
```

```python
print("""******************************
Minik hesap makinesi
******************************""")

islem = input("yapmak istediğin işlemi seç: toplama için 1, çıkarma için 2, bölme için 3 ve çarpma için 4, üs alma için 5e basınız")

sayi1 = int(input("birinci sayıyı gir:"))
sayi2 = int(input("ikinci sayıyı gir:"))

if islem == "1":
    print("{} sayısı ile {} sayısının toplamı {} dır.".format(sayi1, sayi2, sayi1+sayi2))
elif islem == "2":
    print("{} sayısı ile {} sayısının farkı {} dır.".format(sayi1, sayi2, sayi1-sayi2))
elif islem == "3":
    print("{} sayısı ile {} sayısının bölümü {} dır.".format(sayi1, sayi2, sayi1/sayi2))
elif islem == "4":
    print("{} sayısı ile {} sayısının çarpımı {} dır.".format(sayi1, sayi2, sayi1*sayi2))
elif islem == "5":
    print("{} sayısı {} kuvveti  {} dır.".format(sayi1, sayi2, sayi1**sayi2)) 
```

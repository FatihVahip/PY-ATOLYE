```python
print("""*****************
Kullanıcı Girişi
****************""")

s_k_adi = "Kedi"
s_sifre = "123"

can= 3

while True:
    k_adi = input("Kullanıcı adı:")
    sifre= input("Şifre:")
    
    if (s_k_adi !=k_adi and s_sifre==sifre):
        print("Kullanıcı adı hatalı...")
        can -=1
        print("Kalan giriş hakkı:", can)
        
    elif (s_k_adi ==k_adi and s_sifre!=sifre):
        print("Şifren hatalı...")
        can -=1
        print("Kalan giriş hakkı:", can)
        
    elif (s_k_adi !=k_adi and s_sifre!=sifre):
        print("Kullanıcı adı ve şifre hatalı...")
        can -=1
        print("Kalan giriş hakkı:", can)
    else:
        print("Giriş başarılı, tebrikler", k_adi)
        break
    if can ==0:
        print("Giriş hakkın bitti, yarın tekrar bekleriz:)")
        break
```

```python
#ATM programı
print("""İşlemler:
1.Para Yatırma-1'basınız
2.Para Çekme-2'basınız
3.Bakiye Sorgulama-3'basınız
işlemleri sonlandırmak için X basınız
""")

bakiye = 1000
limit = 500

while True:
    islem = input("İşlem seç:")
    
    if islem =="x":
        print("Yine bekleriz, hoşçakal")
        break
    elif islem =="1":
        miktar= int(input("Yatırmak istediğiniz tutarı giriniz:"))
        bakiye = bakiye +miktar
        print("yeni bakiyeniz:", bakiye)
    elif islem== "3":
        print("Hesabınıda bulunan toplam para:", bakiye)
    elif islem =="2":
        miktar = int(input("Çekmek istediğiniz tutarı yazabilir misiniz"))
        if ((bakiye+limit)-miktar<0):
            print("Bakiyeniz ve limitiniz bu kadar para çekmek için uygun değildir")
            print("Bakiyeniz {} tl dir".format(bakiye))
            continue
        bakiye -=miktar
    else:
        print("Geçerli, bir işlem seçiniz....")
```

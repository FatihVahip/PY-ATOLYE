kullanıcıdan şifre ve kullanıcı adı alıp kontrol eden program

```python
k_adi = "Kedi"
sifre = "123"

k_adi = input("Kullanıcı adını gir: ")
sifre = input("Şifreni gir: ")

if k_adi == "Kedi" and sifre == "123":
    print("Hoşgeldin!", k_adi)
    print("Eğlence başlasın......")
elif k_adi == "Kedi" and sifre !="123":
    print("Şifren hatalı", k_adi)
elif k_adi !="Kedi" and sifre=="123":
    print("Kullanıcı adın hatalı!")
else:
    print("Her ikisini de yanlış girdin!") 
```

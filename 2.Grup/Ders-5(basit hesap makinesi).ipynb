```python
print("Toplama için 1 \nÇıkarma için 2 \nBölme için 3 \nÇarpma için 4 \nÜs alma için 5 giriniz")
islem = input("Yapmak istediğin işlemi seç: ")

sayi1 = int(input("Birinci sayıyı gir: "))
sayi2 = int(input("İkinci sayıyı gir: "))

if islem =="1":
    print("{}+{} toplamı: {} ".format(sayi1, sayi2, sayi1+sayi2))
elif islem =="2":
    print("{}-{} Farkı: {} ".format(sayi1, sayi2, sayi1-sayi2))
elif islem =="3":
    print("{}/{} Bölümü: {} ".format(sayi1, sayi2, sayi1/sayi2))
elif islem =="4":
    print("{}*{} Çarpma: {} ".format(sayi1, sayi2, sayi1*sayi2))
elif islem =="5":
    print("{}**{} Üslü hali: {} ".format(sayi1, sayi2, sayi1**sayi2))
else:
    print("Seni şakacı! kaybol:)")
```
